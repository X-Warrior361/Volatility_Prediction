```python
cities_list = pd.read_csv('/kaggle/input/indian-5000-cities-weather-data/Weather_Data_Scraping_and_Analysis/weather.csv', index_col=0)
```

```python
UP_cities = list(cities_list.city.values[[33, 219, 69, 25, 42, 4354, 57, 199, 189, 279]])
MP_cities = list(cities_list.city.values[[398, 454, 1184, 20, 1113, 1095, 347, 103]])
PJ_cities = list(cities_list.city.values[[41, 486, 24, 1084, 65, 183, 60, 1081]])
```

```python
UP_cities = [i for i in UP_cities if i[0] < "S"]
MP_cities = [i for i in MP_cities if i[0] < "S"]
PJ_cities = [i for i in PJ_cities if i[0] < "S"]
```

```python
wd1 = "/kaggle/input/indian-5000-cities-weather-data/w_d_1/"
wd2 = "/kaggle/input/indian-5000-cities-weather-data/w_d_2/"
wd3 = "/kaggle/input/indian-5000-cities-weather-data/w_d_3/w_d_3/"
```

```python
def open_csv(name):
    if name[0] < "H":
        file = pd.read_csv(wd1+name+".csv", index_col=0)
    elif name[0] >= "H" and name[0] < "O":
        file = pd.read_csv(wd2+name+".csv", index_col=0)
    elif name[0] >= "O":
        file = pd.read_csv(wd3+name+".csv", index_col=0)
        
    return file
```

```python
up,mp,pj = [],[],[]

for i in UP_cities:
    up.append(open_csv(i))
    

for j in MP_cities:
    mp.append(open_csv(j))
    

for k in PJ_cities:
    pj.append(open_csv(k))
```

```python
UP = pd.concat(up)
MP = pd.concat(mp)
PJ = pd.concat(pj)
```

```python
def aggregate(df):
    df_new = df.groupby(by='date').mean().reset_index()
    df_new.drop(columns = list(set(df_new.columns)-set(['date','temperature_2m','precipitation'])), inplace = True)
    df_new.columns = ['date', 'temperature', 'precipitation']
    
    dates = [i[:10] for i in df_new.date.values]
    df_new['date'] = dates
    
    df_new_ = df_new.groupby(by='date').mean().reset_index()
    
    return df_new_
```

```python
UP_new = aggregate(UP)
MP_new = aggregate(MP)
PJ_new = aggregate(PJ)
```

```python
UP_new.to_csv('/kaggle/working/uttar_pradesh.csv', index=False)
MP_new.to_csv('/kaggle/working/madhya_pradesh.csv', index=False)
PJ_new.to_csv('/kaggle/working/punjab.csv', index=False)
```
